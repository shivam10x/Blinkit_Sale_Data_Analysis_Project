# Import Libraries____

```python
import pandas as pd
import matplotlib.pyplot as plt 
import seaborn as sns
```

# Load Dataset___

```python
data = pd.read_csv("BlinkIT-Grocery-Data.csv")
data.head()
```

# Data Information___

```python
data.describe()
```

```python
data.info()
```

# Data Cleaning___

```python
# Check Missing Values...

data.isnull().sum()
```

```python
# Fil Or Drop Missing Values...

data.dropna(inplace=True)
```

```python
# Rename Columns ...

data.columns.str.replace(' ' , '_')
```

#                                              Data Visualization___

# 1. What are the most sold product categories?

```python
plt.figure(figsize=(10,6))

sns.countplot(data = data , x = 'Item Type' , palette='Set2')
plt.title('Most Sold Products Categories Based On Sales Count')
plt.xlabel('Categories')
plt.ylabel('Sales Count')
plt.xticks(fontsize = 10 , rotation = 90)
plt.show()
```

Insights : 

Fruits, vegetables, and snacks are usually the most sold

# 2. Which category generates the highest revenue?

```python
highest_revenue = data.groupby('Item Type')['Sales'].sum().sort_values(ascending=False)
sns.set_style('whitegrid')

plt.figure(figsize=(8,6))
sns.barplot(y = highest_revenue.index ,x = highest_revenue.values , palette='Set1')
plt.title('Highest Revenue Based On Category')
plt.xlabel('Category')
plt.ylabel('Count')
plt.show()
```

# 3. How does outlet size affect sales?

```python
Outlet_Size_Sales = data.groupby('Outlet Size')['Sales'].sum().sort_values(ascending=False)

plt.figure(figsize=(10,6))
sns.barplot(x = Outlet_Size_Sales.index , y = Outlet_Size_Sales.values , palette='Set1')
plt.title('Outlet Size Affect Sales')
plt.xlabel('Outlet Size')
plt.ylabel('Count (Sales)')
plt.show()
```

# 4. Does product visibility impact sales?

```python
plt.figure(figsize=(10,6))

sns.scatterplot(data = data , x ='Item Visibility'  , y = 'Sales' , palette='Set1')
```

# 5. Which outlet type performs best?

```python
plt.figure(figsize=(10,6))
sns.barplot(data = data , x = 'Outlet Type' , y = 'Sales' , palette='Set2')
plt.title('Outlet Type Performance')
plt.xlabel('Outlet Types')
plt.ylabel('Count (Sales)')
plt.show()
```

# 6. How do fat content or product types affect sales?

```python
plt.figure(figsize=(10,6))
sns.scatterplot(data = data , x = 'Item Fat Content' , y = 'Sales' , palette='Set1')
plt.title('Fat Content Affect Sales')
plt.xlabel('Fat Content')
plt.ylabel('Count (Sales)')
plt.show()
```

# 7. What is the average sales per outlet?

```python
plt.figure(figsize=(10,6))

sns.boxplot(data = data , x ='Outlet Type' , y = 'Sales' , hue='Outlet Size')
plt.title('Average Sales Per Outlet')
plt.xlabel('Outlet')
plt.ylabel('Count (Sales)')
plt.show()
```
